```python
import numpy as np 
import matplotlib.pyplot as plt
import os
import pandas as pd

path = os.path.expanduser(r"~/Documents/python/identification_modified_proteins/msms_test.ascii")
#os.path.expanduser used to define path for MacOS 

arr = np.loadtxt(path, delimiter=',', dtype=str)

ms_arr = np.array(list(np.core.defchararray.split(arr[8:])), dtype=float)

plt.plot(ms_arr[:,0],ms_arr[:,1])

plt.show()
```

```python
range_splits = list(range(0, 2005, 50))
print(list(range(0, 2005, 50)))
```

```python
split_at = ms_arr[:, 0].searchsorted([0, 50, 100, 150, 200, 250, 300, 350, 400, 450, 500, 550, 600, 650, 700, 750, 800, 850, 900, 950, 1000, 1050, 1100, 1150, 1200, 1250, 1300, 1350, 1400, 1450, 1500, 1550, 1600, 1650, 1700, 1750, 1800, 1850, 1900, 1950, 2000])
#cannot use a variable in .searchsorted([]) as it needs scalar integers. 
arr_split = np.split(ms_arr, split_at)
print(arr_split)
```

```python
intensity_sums = [np.sum(arr[:,1]) for arr in arr_split]
intensity_sums
```

```python
mass_range = np.array(range_splits)
df = pd.DataFrame(data=[mass_range, intensity_sums], index=['Mass Range (Up To)', 'Intensity Sum']).T
#transpose two numpy arrays together to form dataframe
df
```
